## From config

```python
def example_basic_extraction():
    """Basic feature extraction example using minimal configuration."""
    print("=== Basic Feature Extraction Example ===")
    
    # Minimal configuration
    config = {
        'paths': {
            'input_dir': 'data/sample_data',
            'output_dir': 'data/features_output_example'
        },
        'feature_extraction': {
            'method': 'incarta',  # Specify the extraction method
            'n_jobs': 2,  # Use 2 cores for this example
            'output': {
                'save_individual_files': True,
                'save_combined_file': True,
                'include_metadata': True
            },
        },
        'logging': {
            'level': 'INFO'
        }
    }
    
    # Run pipeline
    pipeline = FeatureExtractionPipeline(config=config)
    features_df = pipeline.run()
    
    print(f"Extracted features from {len(features_df)} instances")
    if not features_df.empty:
        print(f"Feature columns: {list(features_df.columns)}")
        print(f"Sample statistics:\n{features_df.describe()}")

```

```python

def example_advanced_extraction():
    """Advanced feature extraction example with custom processing."""
    print("\n=== Advanced Feature Extraction Example ===")
    
    # Advanced configuration
    config = {
        'paths': {
            'input_dir': 'data/sample_data',
            'output_dir': 'data/features_output_advanced'
        },
        'feature_extraction': {
            'method': 'incarta',  # Specify the extraction method
            'n_jobs': -1,
            'features': {
                'morphology': True,
                'intensity': True,
                'spatial': True,
                'texture': True
            },
            'preprocessing': {
                'normalize_intensity': True,
                'clip_percentiles': [1, 99]
            },
            'output': {
                'save_individual_files': True,
                'save_combined_file': True,
                'include_metadata': True,
                'individual_format': '{image_name}_advanced_features.csv',
                'combined_filename': 'all_advanced_features.csv',
                'create_subdirs': True,
            },
        },
        'logging': {
            'level': 'INFO',
            'filename': 'advanced_feature_extraction.log'
        }
    }
    
    # Run pipeline
    pipeline = FeatureExtractionPipeline(config=config)
    features_df = pipeline.run()
    
    print(f"Extracted features from {len(features_df)} instances")
    if not features_df.empty:
        print(f"Number of feature columns: {len(features_df.columns)}")
        
        # Show feature categories
        morphology_cols = [col for col in features_df.columns if col in ['area', 'perimeter', 'elongation', 'compactness', 'circularity']]
        intensity_cols = [col for col in features_df.columns if 'intensity' in col]
        texture_cols = [col for col in features_df.columns if col in ['gabor_mean', 'gabor_std', 'skewness', 'kurtosis', 'entropy']]
        spatial_cols = [col for col in features_df.columns if 'centroid' in col or 'mass' in col]
        
        print(f"Morphology features ({len(morphology_cols)}): {morphology_cols}")
        print(f"Intensity features ({len(intensity_cols)}): {intensity_cols}")
        print(f"Texture features ({len(texture_cols)}): {texture_cols}")
        print(f"Spatial features ({len(spatial_cols)}): {spatial_cols}")
```

```python


if __name__ == "__main__":
    # Run examples
    try:
        example_basic_extraction()
        example_advanced_extraction()
        
        print("\nAll examples completed successfully!")
        print("\nTo run the feature extraction pipeline on your own data:")
        print("1. Update the config file: config/feature_extraction_config.yaml")
        print("2. Run: python scripts/run_feature_extraction.py --config config/feature_extraction_config.yaml")
        print("3. Or use command line options: python scripts/run_feature_extraction.py --input-dir your_data_dir --output-dir your_output_dir")
        
    except Exception as e:
        print(f"Error running examples: {str(e)}")
        print("This is expected if the sample data doesn't exist yet.")
        print("The pipeline is ready to use when you have your data prepared!")
```
